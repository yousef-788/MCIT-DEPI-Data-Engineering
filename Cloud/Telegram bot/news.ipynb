```python
import feedparser
```

```python
url = 'https://news.google.com/rss/search?q=world'
```

```python
data = feedparser.parse(url)
```

```python
data.keys()
```

```python
news = []
for article in data['entries'][0:10]:
    title = article['title']
    link = article['link']
    source = article['source']['title']

    news.append({
        'title': title,
        'link': link,
        'source': source
    })
```
