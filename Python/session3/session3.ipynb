```python
import random
num = random.randint(1,10)
while True :
    guess = int(input("Try to guess"))
    if guess == num :
        print (f"You got it right, {num} is indeed the right number")
        break
    else :
        print(f"{guess} isn't the right number try again")
```

```python
from math import pi
def circle(radius):
    area = pi * pow(radius,2)
    print (f"The area equals to {area:.2f}") 
```

```python
radius = int(input("Enter the radius"))
circle(radius)
```
