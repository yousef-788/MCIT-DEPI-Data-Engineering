```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
 
list_quotes = []
driver = webdriver.Chrome()
 
 
driver.get('http://quotes.toscrape.com/page/1/')
 
while True:
 quotes = driver.find_elements(By.CLASS_NAME, 'quote')
 try:
   for quote in quotes:
    text = quote.find_element(By.CLASS_NAME, 'text').text
    auther = quote.find_element(By.CLASS_NAME, 'author').text
 
    list_quotes.append({
     'text': text,
     'auther': auther
    })
   driver.find_element(By.CSS_SELECTOR, ' div:nth-child(2) > div.col-md-8 > nav > ul > li.next > a').click()
 except:
  break
 
driver.quit()    
 
 
df = pd.DataFrame(list_quotes)
df.to_csv('quotes.csv', index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
 
list_films = []
 
driver = webdriver.Chrome()
 
driver.get('https://www.imdb.com/chart/top/')
 
 
films = driver.find_elements(By.CLASS_NAME, 'ipc-metadata-list-summary-item')
 
for film in films:
    film_name = film.find_element(By.CLASS_NAME, 'ipc-title__text').text
    details = film.find_element(By.CLASS_NAME, 'cli-title-metadata')
    film_year = details.find_element(By.CSS_SELECTOR,'span:nth-child(1)').text
    film_duration = details.find_element(By.CSS_SELECTOR,' span:nth-child(2)').text
    try:
        film_type = details.find_element(By.CSS_SELECTOR,'span:nth-child(3)').text
    except:    
        film_type = 'Null'
    film_rank = film.find_element(By.CLASS_NAME, 'ipc-rating-star--rating').text
    list_films.append({
        'name': film_name,
        'year': film_year,
        'duration': film_duration,
        'type': film_type,
        'rank': film_rank
    })
 
 
driver.quit()
df = pd.DataFrame(list_films)
df.to_csv('Films.csv', index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

list_films = []

# 1. Setup Chrome options for headless mode (runs faster in the background)
chrome_options = Options()
chrome_options.add_argument("--headless") 
chrome_options.add_argument("--window-size=1920,1080")
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")

driver = webdriver.Chrome(options=chrome_options)
driver.get('https://www.imdb.com/chart/top/')

try:
    # 2. Add an Explicit Wait to ensure the elements are loaded before scraping
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, 'ipc-metadata-list-summary-item'))
    )
    
    films = driver.find_elements(By.CLASS_NAME, 'ipc-metadata-list-summary-item')
    
    for film in films:
        raw_name = film.find_element(By.CLASS_NAME, 'ipc-title__text').text
        # 3. Clean the title (IMDb displays titles as "1. The Shawshank Redemption")
        film_name = raw_name.split('. ', 1)[-1] if '. ' in raw_name else raw_name
        
        details = film.find_element(By.CLASS_NAME, 'cli-title-metadata')
        film_year = details.find_element(By.CSS_SELECTOR, 'span:nth-child(1)').text
        film_duration = details.find_element(By.CSS_SELECTOR, 'span:nth-child(2)').text
        
        try:
            # Note: IMDb usually puts Age Rating (e.g., 'R', 'PG-13') in this 3rd span, not genre.
            film_type = details.find_element(By.CSS_SELECTOR, 'span:nth-child(3)').text
        except:    
            film_type = 'Null'
            
        film_rank = film.find_element(By.CLASS_NAME, 'ipc-rating-star--rating').text
        
        list_films.append({
            'name': film_name,
            'year': film_year,
            'duration': film_duration,
            'type': film_type,
            'rank': film_rank
        })

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Always ensure the driver quits even if the script crashes
    driver.quit()

# Save to CSV
df = pd.DataFrame(list_films)
df.to_csv('Films.csv', index=False)
print(f"Successfully scraped {len(df)} movies to Films.csv!")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
from selenium.webdriver.chrome.options import Options
 
jobssearch = ['Python Developer', 'Data Engineer', 'AI Developer', 'Data Scientist']
 
options = Options()
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
 
list_jobs = []
 
driver = webdriver.Chrome(options=options)
 
for i in jobssearch:
 driver.get('https://wuzzuf.net/jobs/egypt')
 
 driver.find_element(By.NAME, 'q').send_keys(i)
 
 driver.find_element(By.XPATH, '//*[@id="app"]/div/div/main/div[2]/div[1]/form/button').click()
 
 
 time.sleep(3)
 
 for i in range(3):
  time.sleep(5)
  jobs = driver.find_elements(By.CLASS_NAME, 'css-pkv5jc')
  for job in jobs:
   
    job_title = job.find_element(By.CLASS_NAME, 'css-o171kl').text
    company = job.find_element(By.CLASS_NAME, 'css-ipsyv7').text
    location = job.find_element(By.CLASS_NAME, 'css-16x61xq').text
    tags = job.find_element(By.CLASS_NAME, 'css-5jhz9n').text.split('/n')
    try:
       exp = job.find_element(By.CSS_SELECTOR, 'div.css-1rhj4yg > div:nth-child(2) > span').text
    except:
      exp = 'No Exp'  
    list_jobs.append({
        'job_title': job_title,
        'company': company,
        'location': location,
        'tags': tags,
        'exp': exp
    })
 
  driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div/div/div[3]/ul/li[7]/button').click()    
 
driver.quit()
df = pd.DataFrame(list_jobs)
df.to_csv('Jobs.csv', index=False)
 
```
