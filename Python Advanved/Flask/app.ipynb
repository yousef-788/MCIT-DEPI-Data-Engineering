```python
from flask import Flask, request, render_template

app = Flask(__name__)

@app.route('/')
def home():
    # جايب البيانات من الـ URL query parameters (GET)
    name = request.args.get('name')
    password = request.args.get('password')

    # بس عشان نشوفهم في الـ terminal
    print(name, password)

    # بنبعتهم للـ HTML
    return render_template('user.html', name=name, password=password)


if __name__ == '__main__':
    app.run(debug=True, use_reloader=False, port=5000)
```

```python
import pandas as pd
def get_Emp(SSN):
    df = pd.read_sql(
        f"select SSN, Fname + ' ' + Lname as Full_Name, Salary, Dname "
        f"from Employee as Emp inner join Departments as dep "
        f"on Emp.Dno = dep.Dnum where SSN = {SSN}",
        conn
    )
    return df
```

```python
from flask import Flask, request, render_template

app = Flask(__name__)

@app.route('/')
def home():
    SSN = request.args.get('SSN')
    df = get_Emp(SSN)
    return render_template("emp.html", tables=df.to_html(classes='table', index=False))


if __name__ == '__main__':
    app.run(debug=True, use_reloader=False, port=5000)
```
