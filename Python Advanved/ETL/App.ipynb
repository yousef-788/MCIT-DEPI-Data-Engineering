### 1. Extract 

```python
import pyodbc
import pandas as pd
```

```python
source_conn = pyodbc.connect(
    "Driver={SQL Server Native Client 11.0};"
    "Server=localhost;"
    "Database=SalesDB;"
    "Trusted_Connection=yes;")
```

```python
df_customer = pd.read_sql("select * from ViewCusomer", source_conn)
df_customer
```

```python
df_product = pd.read_sql("select * from ViewProduct", source_conn)
df_product
```

```python
df_salesman = pd.read_sql("select * from ViewSalesman", source_conn)
df_salesman
```

```python
df_FactOrders = pd.read_sql("select * from ViewFactOrders", source_conn)
df_FactOrders
```

```python
source_conn.close()
```

```python
df_FactOrders['OrderDate'] = pd.to_datetime(df_FactOrders['OrderDate'])
```

```python
df_FactOrders['OrderDate'] = df_FactOrders['OrderDate'].astype(int)
df_FactOrders['Quantity'] = df_FactOrders['Quantity'].astype(float)
```

### 3. Load

```python
dwh_conn = pyodbc.connect(
    "Driver={SQL Server Native Client 11.0};"
    "Server=localhost;"
    "Database=SalesDWH;"
    "Trusted_Connection=yes;")
```

```python
cusror = dwh_conn.cursor()
```

```python
df_customer
```

```python
for index, row in df_customer.iterrows():
    cusror.execute('insert into DimCustomer values (?,?,?,?,?)',row['CustomerID'],row['FullName'],row['PhoneNumber'],row['Email'],row['GenderName'])
dwh_conn.commit()    
```

```python
df_product
```

```python
for index, row in df_product.iterrows():
    cusror.execute('insert into DimProduct values (?,?,?,?,?)',row['ProductID'],row['ProductName'],row['Price'],row['SubCategoryName'],row['CategoryName'])
dwh_conn.commit()    
```

```python
df_salesman
```

```python
for index, row in df_salesman.iterrows():
    cusror.execute('insert into DimSalesman values (?,?,?,?,?)',row['SalesmanID'],row['FullName'],row['Email'],row['PhoneNumber'],row['City'])
dwh_conn.commit()    
```

```python
df_FactOrders.head(5)
```

```python
for index, row in df_FactOrders.iterrows():
    cusror.execute('insert into FactOrders values (?,?,?,?,?,?,?)',row['OrderID'],row['CustomerID'],row['SalesmanID'],row['OrderDate'],row['ProductID'],row['Quantity'],row['TotalPrice'])
dwh_conn.commit()    
```
