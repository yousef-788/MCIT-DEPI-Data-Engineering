```python
import math
radius = int (input ("enter the radias of the circle "))

area = abs(radius)**2*math.pi

print (f"The area is {area}")
```

```python
Num1 = int (input("Enter First Number"))
Num2 = int (input("Enter Second Number"))
OP = input("Enter The Operatop")

if OP == '+' :
    print (f"{Num1} + {Num2} = {Num1+Num2}")
elif OP == '-' :
    print (f"{Num1} - {Num2} = {Num1-Num2}")
elif OP == '*' :
    print (f"{Num1} x {Num2} = {Num1*Num2}")
elif OP == '/' :
    print (f"{Num1} / {Num2} = {Num1/Num2}")
else :
    print("This isn't a valid operator")
     
```

```python
import time

noOFseconds = int(input("Enter The Number Of Seconds: "))

for i in range(noOFseconds, -1, -1):
    print(f"Time left: {i} seconds")
    time.sleep(1)

print("Time's up!")
```
