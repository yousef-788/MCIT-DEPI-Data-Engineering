```python
products={
"pizza":10,
"shawrma":5,
"meal":15,
"cola":8}
```

```python
for k,v in products.items():
    print(f"{k}:{v}")
```

```python
total = 0
while True:
    product = input("Enter the product name")
    if product == "exit":
        break
    elif product in products :
        print(f"The price of {product} is {products[product]}",end="  ")
        total+=products[product]
        print(f"the total is {total}")
    else:
        print("Product not found")
```
