```python
def deposit(balance, money):
    balance += money
    print("successful transaction")
    print(f"the available balance is {balance}\n") 
    return balance
```

```python
def withdraw(balance, money):
    if money <= balance:
        balance -= money
        print("successful transaction")
    else:
        print("cannot withdraw more than the available balance")

    print(f"the available balance is {balance}\n")
    return balance
```

```python
def show_balance(balance):
    print(f"the available balance is {balance}\n") 
    return balance
```

```python
def main(balance):
    while True:
        choice = int(input("available options: 1) Deposit 2) Withdraw 3) Show Balance 4) Exit\n"))

        if choice == 1:
            money = int(input("How much do you want to deposit?\n"))
            if money > 0:
                balance = deposit(balance, money)
            else:
                print("Enter a valid amount of money\n")

        elif choice == 2:
            money = int(input("How much do you want to withdraw?\n"))
            if money > 0:
                balance = withdraw(balance, money)
            else:
                print("Enter a valid amount of money\n")

        elif choice == 3:
            balance = show_balance(balance)

        elif choice == 4:
            break

        else:
            print("Enter a valid choice number\n")

    return balance
```

```python
balance = 0
main(balance)
```
