```python
products = []
prices = []

while True:
    product = input("Enter the item: ")
    if product == "q":
        break
    
    products.append(product)
    
    price = int(input("Enter the price: "))
    prices.append(price)

print(products)
print(prices)

total = 0
for i in prices:
    total += i

print(total)
```

```python
products = {
    "flour":10,
    "suger":20,
    "milk":30,
    "bread":40
}
for k,v in products.items() :
    print(f"{k}:{v}")

cart = []
while True :
    item = input ("Enter the name of your item")
    if item.lower() == "q":
        break
    if item in products:
        cart.append(products[item])
    else:
        print("Item not found!")
print(cart)
total=0
for i in cart:
    total+=i
print(total)
    
    
```
