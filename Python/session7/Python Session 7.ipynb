```python
student = {
    "Names": ["Ali", "Ahmed", "Khalid" , "Mona", "Fatma"],
    "Age" : [20, 21, 22, 20 , None],
    "Gender" : ["Male", "Male", "Male" ,"Female", "Female"],
    "Dept_ID" : [1, None, 1, 3 , None]
}

departments = {
    "Dept_ID" : [1, 2, 3 , 4],
    "Dept_Name" : ["HR", "IT", "Finance", "Call Center"],
    "Dept Manager" : ["Mohamed", "Sayed", "Salma" , "Mina"]
}
```

```python
df_students = pd.DataFrame(student)
df_departments = pd.DataFrame(departments)
```
