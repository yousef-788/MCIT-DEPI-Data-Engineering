```python
class Hospital:
    def __init__(self, name, location):
        self.name=name
        self.location=location
        self.doctors=[]
        self.patients=[]

    def add_doctors(self, *doctor):
        self.doctors.extend(doctor)
    
    def add_patients(self, *patient):
        self.patients.extend(patient)
        
    def show_doctors(self):
        print(", ".join(self.doctors))

    def show_patients(self):
        print(", ".join(self.patients))
    
```

```python
hospital1=Hospital("El Salam", "Nasr City, Cairo")
hospital1.add_doctors("mohamed","ahmed")
hospital1.show_doctors()
hospital1.add_patients("ali","ibrahem")
hospital1.show_patients()
```

```python
class Doctor(Hospital):
    def __init__(self, name, specialty):
        self.name=name
        self.specialty=specialty
        self.patients=[]
    def assign_patient(self, *patient):
        self.patients.extend(patient)
    
    def show_patients(self):
        print(", ".join(self.patients))

```

```python
doctor1=Doctor("mohamed","Pediatrician")
doctor1.assign_patient("mohamed", "ahmed")
doctor1.show_patients()
```

```python
class Patient(Hospital):
    def __init__(self, name, age, illness):
        self.name=name
        self.age=age
        self.illness=illness
        
    def  assign_doctor(self, doctor):
        self.doctor=doctor
        
    def  show_info(self):
        print(f"""patient's name is {self.name} 
patient's age is {self.age} 
the patient's has {self.illness}
his doctor is {self.doctor}
""")
```

```python
patient1=Patient("ahmed",25,"fever")
patient1.assign_doctor("ali")
patient1.show_info()
```
