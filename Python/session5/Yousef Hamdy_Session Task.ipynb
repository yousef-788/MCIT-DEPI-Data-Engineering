```python
with open("egypt.txt","r") as object :
    data = object.read()
    print(data)
```

```python
cities = {}

for line in data.splitlines():
    parts = line.split(",")
    
    city = parts[0]
    
    if len(parts)>1 and not parts[1].isdigit():
        city_type = parts[1]
    else:
        city_type="Unknown"
        
    city_pop=0
    for part in parts :     
        if part.isdigit():
            city_pop = int(part) 
    
    cities[city] = {
        "type" : city_type,
        "population" : city_pop 
    }

for key, value in cities.items():
    print(f"{key} : {value}")
```

```python
sorted_cities = sorted(cities.items(),key = lambda item : item[1]["population"])
sorted(cities.items(),key = lambda item : item[1]["population"])
```

```python
print(f"The most populated city is {sorted_cities[-1][0]}")
```

```python
print(f"The least populated city is {sorted_cities[0][0]}")
```

```python
import json
with open("data.json" , "w") as object:
   json.dump(cities, object , indent=4)
```

```python
while True:
    city_chosen = input("Choose a city or enter '0' to exit: ")

    if city_chosen == "0":
        break

    elif city_chosen in cities:
        print(cities[city_chosen])

    else:
        print("Not found"
```
