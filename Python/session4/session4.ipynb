```python
class Bank_account:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance
    
    
    def deposit(self, money):
        self.balance += money
        print("successful transaction")
        print(f"the available balance is {self.balance}\n") 
        return self.balance
    
    def withdraw(self, money):
        if money <= self.balance:
            self.balance -= money
            print("successful transaction")
        else:
            print("cannot withdraw more than the available balance")

        print(f"the available balance is {self.balance}\n")
        return self.balance
    
    def show_balance(self):
        print(f"the available balance is {self.balance}\n") 
        return self.balance
```

```python
account1 = Bank_account("ahmed",6000)
account1.show_balance()
account1.deposit(5000)
account1.withdraw(3000)
```

```python
from math import pi

class Shape:
    def __init__(self, color):
        self.color=color
        
class Rectangle(Shape):
    def __init__(self, color, width, hight):
        super().__init__(color)
        self.width=width
        self.hight=hight
    def get_area(self):
        area = self.width*self.hight
        print(f"{area:.2f}")
    
class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius=radius
        
    def get_area(self):
        area =  pi*pow(self.radius,2)
        print(f"{area:.2f}")
```

```python
rect=Rectangle("red",5,10)
rect.get_area()

cir=Circle("red",10)
cir.get_area()
```
