```python
import pandas as pd
df = pd.read_csv("data.csv")
```

```python
df.duplicated().sum()
```

```python
df["Price per Unit"] = df["Price per Unit"].str.replace("$", "")
df["Price per Unit"] = df["Price per Unit"].astype(float)
df["Operating Profit"] = df["Operating Profit"].str.replace("$", "").str.replace(",", "").str.strip()
df["Operating Profit"] = df["Operating Profit"].astype(float)
```

```python
df.isnull().sum()
```

```python
df = df.fillna({
    "Region" : df["Region"].mode()[0],
    "Price per Unit": df["Price per Unit"].mean()             
})

df.isnull().sum()
```

```python
df["Invoice Date"] = pd.to_datetime(df["Invoice Date"])

df["year"] = df["Invoice Date"].dt.year
df["month"] = df["Invoice Date"].dt.month
df["day"] = df["Invoice Date"].dt.day
```

```python
df.drop(columns="Retailer ID" , inplace=True)
```

```python
print(df["Region"].unique())
df["Region"] = df["Region"].str.title() 
print(df["Region"].unique())
```

```python
df["Total Sales"] = df["Total Sales"].str.replace(",","")
df["Total Sales"] = df["Total Sales"].astype(float)
total_sales = df["Total Sales"].sum()
print(f"total sales is {int(total_sales)}")
```

```python
df.groupby("Region")["Total Sales"].mean()
```

```python
df.groupby("Product")["Total Sales"].mean()
```

```python
df.to_csv("modified_data.csv")
```
